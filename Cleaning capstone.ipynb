```python
EUdivorce= pd.read_excel(r'C:\Users\sh9wn\Desktop\divorce statistics euro.xlsx', sheet_name='Sheet2')
```

```python
EUdivorce = EUdivorce.iloc[5:]
#drop non useful rows
```

```python
EUdivorce.set_index('country_name',inplace=True)
#index set to country name for easy navigation
```

```python
EUdivorce.reset_index('country_name',inplace=True)
```

```python
EUdivorce.country_name = EUdivorce.country_name.apply(lambda x: x.split(' ')[0])
#remove extra wording from Germany and Kosovo
```

```python
EUdivorce.set_index('country_name',inplace=True)
```

```python
EUdivorce.rename(index={'United':'United Kingdom', 'North':'Macedonia', 'Czechia': 'Czech Republic'},inplace=True)
```

```python
EUdivorce['mean']= EUdivorce.iloc[:][EUdivorce.columns].mean(axis=1)
#created avg of all years for each country
```
