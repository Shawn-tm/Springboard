```python
EUdivorce= pd.read_excel(r'C:\Users\sh9wn\Desktop\divorce statistics euro.xlsx', sheet_name='Sheet2')
```

```python
EUdivorce
#! use jupyter lab
```

```python
#drop error columns
EUdivorce.drop(columns=['Unnamed: 2','Unnamed: 4','Unnamed: 6','Unnamed: 8','Unnamed: 10','Unnamed: 12','Unnamed: 14','Unnamed: 16','Unnamed: 18','Unnamed: 20','Unnamed: 22','Unnamed: 24'], inplace=True)
```

```python
EUdivorce = EUdivorce.iloc[5:]
#drop non useful rows
```

```python
EUdivorce= EUdivorce.rename(columns={'time':'country_name'})
```

```python
EUdivorce.set_index('country_name',inplace=True)
#index set to country name for easy navigation
```

```python
EUdivorce= EUdivorce.apply(lambda x: x.replace(':',np.nan))
#replace : with nan
```

```python
EUdivorce.dropna(thresh=9,inplace=True)
#drop results with more than 3 nan's
```

```python
EUdivorce.fillna(axis=1,method='ffill',inplace=True)
#used ffill from last valid year in row
```

```python
EUdivorce= EUdivorce.astype(float)
```

```python
EUdivorce.reset_index('country_name',inplace=True)
```

```python
EUdivorce.country_name = EUdivorce.country_name.apply(lambda x: x.split(' ')[0])
#remove extra wording from Germany and Kosovo
```

```python
EUdivorce.set_index('country_name',inplace=True)
```

```python
EUdivorce.rename(index={'United':'United Kingdom', 'North':'Macedonia', 'Czechia': 'Czech Republic'},inplace=True)
```

```python
EUdivorce['mean']= EUdivorce.iloc[:][EUdivorce.columns].mean(axis=1)
#created avg of all years for each country
```
