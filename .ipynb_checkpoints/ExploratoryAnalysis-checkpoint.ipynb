```python
eurocide= suicideDf[suicideDf['country'].isin(europels)]
eurocide.head()
```

```python
eurocide[['country','age','year','suicides/100k pop']].nlargest(10,columns='suicides/100k pop')
#top 10 results. Hungary has 7 
```

```python
eurocide.info()
```

```python
totaldeath_byage= eurocide.groupby('age')[['suicides_no']].sum().reindex(['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'])
totaldeath_byage.reset_index(inplace=True)
```

```python
totaldeath_byage['suicides_no']= totaldeath_byage['suicides_no'].apply(lambda x: x/1000)
```

```python
totaldeath_byage.age = totaldeath_byage.age.apply(lambda x: x.split(' ').pop(0))
```

```python
totaldeath_byage
```

```python
plt.figure(figsize=(10,8))
plt.rc('font',size=20)
plt.bar(x='age', height='suicides_no', data= totaldeath_byage, color=['b','b','b','r','b','b'])
plt.xlabel('Age Group (years)')
plt.ylabel('Total Suicides (thousands)')
plt.title('European Total Suicides by Age Group')
plt.xticks(rotation=45)
plt.tight_layout()
#age 35-54 has the highest count over the years in Europe
```

```python
yearavg_byage= eurocide.groupby('age')[['suicides/100k pop']].mean().reindex(['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'])
yearavg_byage.reset_index(inplace=True)
```

```python
yearavg_byage.age = yearavg_byage.age.apply(lambda x: x.split(' ').pop(0))
```

```python
plt.figure(figsize=(10,8))
plt.rc('font',size=20)
plt.bar(x='age', height='suicides/100k pop', data= yearavg_byage, color=['b','b','b','b','b','r'])
plt.xlabel('Age Group (years)')
plt.ylabel('Avg. Suicides')
plt.title('European Avg. Suicides per 100k People')
plt.xticks(rotation=45)
plt.tight_layout()
#75+ has highest yearly avg of suicides
```

```python
meanby_yearsex.reset_index(inplace=True)
```

```python


sns.lmplot( data=meanby_yearsex.reset_index().iloc[:63],x='year',y='mean100kyear',hue='sex', height=8, aspect=12/8)
plt.title('Mean Annual Suicides')
plt.xlabel('Year')
plt.ylabel('Mean suicides per 100k population')

#males are much more likely to commit suicide, with more variance
```

```python
sns.set(font_scale=1)
sns.pairplot(suicidevlat)
#all of the variables seem to have a positive correlation with each other, especially divorce rate w/ suicides
```

```python
plt.figure(figsize=(16,12))
sns.set(font_scale=2)
p1= sns.scatterplot(x=suicidevlat['latitude'],y=suicidevlat['suicides/100k pop'], hue=suicidevlat['country'],s=500, legend=None)

sns.set(font_scale=0.8)
for line in range(0,suicidevlat.shape[0]):
     p1.text(suicidevlat.latitude[line]+0.01, suicidevlat['suicides/100k pop'][line], 
     suicidevlat.country[line], horizontalalignment='left', 
     size='medium', color='black', weight='semibold')



sns.set(font_scale=2)

plt.xlabel('Latitude')
plt.ylabel('Suicides per 100k population')
plt.title('Suicides by Latitude')
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(16,12))
p2= sns.scatterplot(x=suicidevlat['mean divorce rate/1000'],y=suicidevlat['suicides/100k pop'], hue=suicidevlat['country'],s=500, legend=None)


sns.set(font_scale=0.8)
for line in range(0,suicidevlat.shape[0]):
     p2.text(suicidevlat['mean divorce rate/1000'][line]+0.01, suicidevlat['suicides/100k pop'][line], 
     suicidevlat.country[line], horizontalalignment='left', 
     size='medium', color='black', weight='semibold')

sns.set(font_scale=2)
plt.xlabel('Mean Divorce Rate per 1000 (crude)')
plt.ylabel('Suicides per 100k (crude)')
plt.title('Suicides by Divorce Rate')
```

```python
regiondict= { 'Scandanavia' : ['Sweden', 'Norway', 'Finland', 'Denmark'], 'Western Europe' : ['United Kingdom', 'Switzerland', 'France', 'Belgium', 'Netherlands', 'Ireland', 'Austria', 'Germany','Portugal', 'Spain', 'Italy', 'Luxembourg'], 'Eastern Europe' : ['Ukraine', 'Slovenia', 'Hungary', 'Slovakia', 'Serbia', 'Poland', 'Romania', 'Lithuania', 'Latvia', 'Estonia', 'Croatia', 'Czech Republic','Bulgaria', 'Belarus'],  'Mediterranean': ['Georgia','Turkey', 'Greece','Montenegro', 'Azerbaijan']}
```

```python
reversed_dict = {val: key for key in regiondict for val in regiondict[key]}
```

```python
suicidevlat['Region'] =suicidevlat['country'].map(reversed_dict)
```

```python
suicidevlat.head()
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(16,12))
p2= sns.scatterplot(x=suicidevlat['mean divorce rate/1000'],y=suicidevlat['suicides/100k pop'], hue=suicidevlat['Region'],size=suicidevlat['latitude'], sizes= (100, 1000) )
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1, frameon=True, labelspacing=1.5)


sns.set(font_scale=0.8)
for line in range(0,suicidevlat.shape[0]):
     p2.text(suicidevlat['mean divorce rate/1000'][line]+0.01, suicidevlat['suicides/100k pop'][line], 
     suicidevlat.country[line], horizontalalignment='left', 
     size='medium', color='black', weight='semibold')

sns.set(font_scale=2)
plt.xlabel('Mean Divorce Rate per 1000 (crude)')
plt.ylabel('Suicides per 100k (crude)')
plt.title('Suicides by Divorce Rate')
```

```python
eurocide_others= eurocidetemp2[eurocidetemp2['Mobile cellular subscriptions (per 100 people)'].notna()]
```

```python

def yearplot(year):
    data= eurocide_others[eurocide_others.year== year]
    data.groupby('country', as_index=False)[['Mobile cellular subscriptions (per 100 people)','suicides/100k pop']].mean().plot.scatter('Mobile cellular subscriptions (per 100 people)','suicides/100k pop')
    
    plt.axis(ymin=0,ymax=45,xmin=0, xmax=180)
```

```python
from IPython.display import display
from ipywidgets import interact, widgets
```

```python
v= interact(yearplot, year=widgets.IntSlider(min=2000, max=2015, step=1, value=2000))
display(v)
```
