```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as pil
import keras as kr
import tensorflow as tf
```

```python
#if we use a GPU
def device(path):
    return tf.device(path)
```

```python
#loads and automates image resizing so our image isn't too big
im_size= 512
def img_load(file):
    image= pil.open(file)
    #resizing the image
    factor = im_size/ np.max(image.size)
    size = np.array(image.size) * factor
    #help reduce pixelation from resizing
    size = size.astype(int)
    image = image.resize(size, pil.LANCZOS)
    return np.float32(image)
```

```python
plt.figure(figsize=(12,12))

content_path=r'C:\Users\sh9wn\Desktop\Springboard\ST_pics\nyc.jpg'
style_path=r'C:\Users\sh9wn\Desktop\Springboard\ST_pics\abstract.jpg'

content = img_load(content_path).astype('uint8')
style = img_load(style_path).astype('uint8')

plt.subplot(1, 2, 1)
plt.imshow(content)

plt.subplot(1, 2, 2)
plt.imshow(style)
plt.show()
```
