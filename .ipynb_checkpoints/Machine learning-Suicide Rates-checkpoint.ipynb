```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('precision',3)
#shows only 3 decimals

%matplotlib inline
```

```python
eu_ml_df= eurocide_others.groupby(['country','year'])[eurocide_others.columns[2:]].mean().reset_index()
eu_ml_df.head()
```

```python
keys= EUdivorce.columns[1:13]
```

```python
Divorce_melt = pd.melt(EUdivorce, id_vars='country', value_vars=keys, var_name='year', value_name='divorce_rate')
Divorce_melt.head()
```

```python
Divorce_melt.year= Divorce_melt.year.astype('int64')
```

```python
eu_df = pd.merge(eu_ml_df, Divorce_melt, on=['country','year'])
eu_df.head()
```

```python
eutemp = EUdivorce.drop(EUdivorce.columns[1:14], axis=1)
```

```python
eu_df= pd.merge(eu_df, eutemp, how='left', on='country')
eu_df.head()
```

```python
eu_df = eu_df.rename(columns={'suicides/100k pop': 'suicides_100k', 'Mobile cellular subscriptions (per 100 people)': 'Mobile_subs_100', 'Population density (people per sq. km of land area)': 'pop_density_perkm', 'gdp_per_capita ($)' : 'gdp_pc', 'Electric power consumption (kWh per capita)': 'Electric_pc_pc', 'GNI per capita, PPP (current international $)': 'GNI_per_capita'})
eu_df.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
m= ols('suicides_100k ~ divorce_rate', eu_df).fit()
m.summary()
```

```python
bigm = ols('suicides_100k ~ Mobile_subs_100 + year + divorce_rate + latitude + gdp_pc + population', eu_df).fit()
bigm.summary()
```

```python
sm.qqplot(bigm.resid, line='r')
plt.title('OLS QQ Plot')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(bigm, ax=ax, criterion="cooks", alpha=0.05)
plt.title('Influence Plot')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import explained_variance_score, mean_squared_error

X = eu_df.drop(['suicides_100k','suicides_no','country','population'], axis=1)
y= eu_df.suicides_100k

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("R^2: {}".format(lr.score(X_test, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test,y_pred)))
print('explained var: {}'.format(explained_variance_score(y_test, y_pred)))
```

```python
lm_df= pd.DataFrame({'columns': X.columns, 'coefficient' : lr.coef_})
lm_df.rename(columns={'columns':'col'}, inplace=True)
lm_df
```

```python
df_skl = pd.DataFrame({'Predicted': y_pred,'Actual': y_test})

g = sns.lmplot(data=df_skl, y='Actual', x='Predicted')
g.fig.set_size_inches(8,8)
plt.title('Suicides per 100k')
#transpose xy
```

```python
from sklearn.linear_model import Lasso, LassoCV
alphas=[0.0001,0.001,0.01, 0.1, 1]
for i in alphas:
    lasso = Lasso(alpha=i, normalize=True, random_state=42)
    lasso.fit(X_train, y_train)
    y_pred = lasso.predict(X_test)
    print("R^2 for {}: {}".format(i, lasso.score(X_test, y_test)))
    print('MSE: {}'.format(mean_squared_error(y_test,y_pred)))
#tuing for alpha in Lasso slightly increases R2
#!add variables
```

```python
lassocv= Lasso(alpha= 0.001, normalize=True, random_state=42)
lassocv.fit(X_train, y_train)
y_pred = lassocv.predict(X_test)
print("R^2: {}".format(lassocv.score(X_test, y_test)))
```

```python
# Import necessary modules
from sklearn.linear_model import Ridge

alph=[0.0001, 0.001,0.01]
for i in alph:
    ridge = Ridge(alpha=i, normalize=True, random_state=42)
    ridge.fit(X_train, y_train)
    y_pred = ridge.predict(X_test)
    print("R^2 for {}: {}".format(i, ridge.score(X_test, y_test)))
    print('MSE: {}'.format(mean_squared_error(y_test,y_pred)))
```

```python
ridge = Ridge(alpha=0.001, normalize=True, random_state=42)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
print("R^2 for : {}".format(ridge.score(X_test, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

train_error = np.empty(5)
test_error = np.empty(5)
for degree in range(1,5):
    
    est = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    est.fit(X_train, y_train)
    train_error[degree] = mean_squared_error(y_train, est.predict(X_train))
    test_error[degree] = mean_squared_error(y_test, est.predict(X_test))
    
plt.plot(np.arange(5), train_error, color='green', label='train')
plt.plot(np.arange(5), test_error, color='red', label='test')
plt.ylim(0,60)
plt.ylabel('mse')
plt.xlabel('degree')
plt.legend(loc='lower left')
```

```python
ridge_best = make_pipeline(PolynomialFeatures(2),Ridge(alpha=0.001, normalize=True, random_state=42))
ridge_best.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)
print("R^2 : {}".format(ridge.score(X_test, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test,ridge_pred)))
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
param_grid= {"alpha": np.arange(.001,.01,.001),
                      "l1_ratio": np.arange(0.0, 1.0, 0.01)}
elasto = ElasticNet()
gs_cv = GridSearchCV(elasto, param_grid, cv=5)
gs_cv.fit(X_train, y_train)
grid_pred = gs_cv.predict(X_test)
print("Tuned ElasticNet params: {}".format(gs_cv.best_params_))
print('R2: {}'.format(explained_variance_score(y_test, grid_pred)))
print('MSE: {}'.format(mean_squared_error(y_test,grid_pred)))
#ElasticNet did't improve R2 at all.
```

```python
#pdp plot : sklearn inspection
from sklearn.inspection.partial_dependence import partial_dependence, plot_partial_dependence
fig, ax = plt.subplots(figsize=(15,15))
plot_partial_dependence(lassocv, X=X, features=[0,1,6,16, 17, 18], fig=fig, 
                        feature_names=X.columns)
```

```python
dat= {'Standard': lr.score(X_test,y_test), 'Lasso': lassocv.score(X_test, y_test), 'Ridge': ridge.score(X_test, y_test),
 "Elastic Net" : explained_variance_score(y_test, grid_pred)}
result_df = pd.Series(dat).to_frame('R2 val')
result_df
```

```python
plt.figure(figsize=(7,10))
result_df.plot(kind='bar', legend=False)

plt.ylim(0.5,0.6)
plt.title('R2 w/ Different Reg. Normalization')
plt.xlabel('Model Type')
plt.ylabel('R2 score')
#elastic net scored slight lower than lasso
```

```python
from sklearn.ensemble import RandomForestRegressor

n_estimators= [10,100, 300, 1000]
r2s=[]
MSE=[]
for i in n_estimators:
    rf_reg = RandomForestRegressor(n_estimators=i,random_state=42,oob_score = True )
    rf_reg.fit(X_train, y_train)
    rf_pred = rf_reg.predict(X_test)
    r2s.append(rf_reg.score(X_test, y_test))
    MSE.append(mean_squared_error(y_test,rf_pred))
    #Best R2 is 100
    
```

```python

plt.figure(figsize=(8,6))
plt.plot(n_estimators,r2s, '.-', ms=20, label='r2 score')
plt.plot(n_estimators, MSE, '.-', ms=20, label='MSE')
plt.ylim(0,3.5)
plt.xlabel('Number of Estimators')
plt.ylabel('Score')
plt.legend()
```

```python
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True)
rf_reg.fit(X_train, y_train)
y_train_predicted = rf_reg.predict(X_train)
y_test_predicted_full_trees = rf_reg.predict(X_test)
mse_train = mean_squared_error(y_train, y_train_predicted)
mse_test = mean_squared_error(y_test, y_test_predicted_full_trees)
print("RF with full trees, Train MSE: {} Test MSE: {}".format(mse_train, mse_test))
```

```python
rf_prune = RandomForestRegressor(n_estimators=100, min_samples_leaf=25,random_state=42 )
rf_prune.fit(X_train, y_train)
y_train_predicted = rf_prune.predict(X_train)
y_test_predicted_pruned_trees = rf_prune.predict(X_test)
mse_train2 = mean_squared_error(y_train, y_train_predicted)
mse_test2 = mean_squared_error(y_test, y_test_predicted_pruned_trees)
print("RF with pruned trees, Train MSE: {} Test MSE: {}".format(mse_train2, mse_test2))
```

```python
compare= {'Train data': mse_train, 'Test data': mse_test, 'Diff': mse_test-mse_train}
compare_df = pd.Series(compare).to_frame('MSE Unpruned')

compare_df.plot(kind='bar', legend=None)
plt.ylabel('MSE')
plt.title('Unpruned RF Model')
```

```python
compare2= {'Train data': mse_train2, 'Test data': mse_test2}
compare_df2 = pd.Series(compare2).to_frame('MSE Unpruned')

compare_df2.plot(kind='bar', legend=None, color='r')
plt.ylabel('MSE')
plt.title('Pruned RF Model')
```

```python
mse_test_ls=[]
mse_train_ls=[]
scores=[]
rang= range(1,100)
for i in rang:
    rf=RandomForestRegressor(n_estimators=i, random_state=42)
    rf.fit(X_train, y_train)
    y_train_predicted = rf.predict(X_train)
    y_test_predicted = rf.predict(X_test)
    mse_train = mean_squared_error(y_train, y_train_predicted)
    score = rf.score(X_test, y_test)
    mse_test = mean_squared_error(y_test, y_test_predicted)
    mse_test_ls.append(mse_test)
    mse_train_ls.append(mse_train)
    scores.append(score)

plt.figure(figsize=(8,6))
plt.plot(rang, mse_train_ls, label='Train MSE')
plt.plot(rang, mse_test_ls, label='Test MSE')
plt.xlabel('# of Trees')
plt.ylabel('MSE')
plt.ylim(0,7)
plt.legend()
```

```python
plt.figure(figsize=(8,6))
plt.plot(rang, scores)
plt.xlabel('# of trees')
plt.ylabel('R2 Score')
```

```python
param_grid = {
    'bootstrap': [False, True],
    'max_depth': [100],
    'max_features': [5],
    'min_samples_leaf': [1],
    'min_samples_split': [2,3],
    'n_estimators': [300]
}

rfg= RandomForestRegressor()
gs= GridSearchCV(rfg, param_grid=param_grid, n_jobs=-1)
gs.fit(X_train, y_train)
gs.best_params_
```

```python
gspred= gs.predict(X_test)
explained_variance_score(y_test, gspred)
```

```python
rf_best= RandomForestRegressor(bootstrap=False, max_depth=100, max_features=5, min_samples_leaf=1, min_samples_split=3,
                              n_estimators=300)
rf_best.fit(X_train, y_train)
best_pred = rf_best.predict(X_test)
print("R^2 : {}".format(rf_best.score(X_test, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test,best_pred)))
rf_best.feature_importances_
rf_best.score(X_train, y_train)
```

```python
y_train_predicted = rf_best.predict(X_train)
y_test_predicted_pruned_trees = rf_best.predict(X_test)
mse_train2 = mean_squared_error(y_train, y_train_predicted)
mse_test2 = mean_squared_error(y_test, y_test_predicted_pruned_trees)
print("RF with high performance model, Train MSE: {} Test MSE: {}".format(mse_train2, mse_test2))
```

```python
compa= {'Train data': mse_train2, 'Test data': mse_test2, "Diff betw. Train & Test": (mse_test2-mse_train2)}
compa_df = pd.Series(compa).to_frame('MSE Unpruned')

compa_df.plot(kind='bar', legend=None)
plt.ylabel('MSE')
plt.title('Optimal RF Model')
plt.ylim(0,2.5)
```
